```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("Resources/cards.png")
```

```python
# [365, 348], [574, 338], [402, 659], [642, 643]
# Got the pixel values using paint app and checking for each corner of a card

pts1 = np.float32([[365, 348], [574, 338], [402, 659], [642, 643]])
print(pts1)
```

```python
width, height = 250, 350
pts2 = np.float32([[0,0], [width, 0], [0, height], [width, height]])
```

```python
matrix = cv2.getPerspectiveTransform(pts1, pts2)
outputImg = cv2.warpPerspective(img, matrix, (width, height))
```

```python
for x in range (0, 4):
    cv2.circle(img, (int(pts1[x][0]), int(pts1[x][1])), 5, (0, 0, 255), cv2.FILLED)
```

```python
plt.figure(figsize = (10, 10))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB));
```

```python
plt.figure(figsize = (10, 10))
plt.imshow(cv2.cvtColor(outputImg, cv2.COLOR_BGR2RGB));
```
